```julia
using Polynomials, JLD
```

```julia
setprecision(BigFloat,2048)
```

```julia
function f(z; λ=1, b=2)
    return λ*z*(1+z*(b+z));
end
```

```julia
function df(z; λ=1, b=2)
    return λ*(z*(2*b + 3*z) + 1)
end
```

```julia
cp(b) = -(b+sqrt(b^2-3))/3
```

```julia
k=7
q=2^k
```

```julia
g128 = load("g128.jld")["g128"]
```

```julia
a0 = g128[q+1]
```

```julia
α0 = (-2*BigFloat(pi)*im/a0)^(1/q)
```

```julia
function fq(z; λ=1, b=2)
    for i=1:q
        z = f(z; λ=λ, b=b)
    end
    return z
end
```

```julia
function dfq(z; λ=1, b=2)
    dz=1
    for i=1:q
        dz *= df(z; λ=λ, b=b)
        z = f(z; λ=λ, b=b)
    end
    return z, dz
end
```

```julia
function isPeriodQ(z; λ=1, b=2)
    return fq(z; λ=λ, b=b)-z
end
```

```julia
function GetPeriodicPoint(δ,b; z=α0*δ, numiter=100)
    h = BigFloat(1e-200)
    λ = exp(2*BigFloat(pi)*im*(1/q+δ^q))
    for i=1:numiter
        #w, dw = dfq(z; λ=λ, b=b)
        #z -= w/dw
        w0 = isPeriodQ(z; λ=λ, b=b)
        w1 = isPeriodQ(z+h; λ=λ, b=b)
        w2 = isPeriodQ(z-h; λ=λ, b=b)
        z -= 2*w0*h/(w1-w2)
    end
    return z
end
```

```julia
δ = 1e-2
x = GetPeriodicPoint(δ,2)
```

```julia
function H(δ, b, x0)
    λ=exp(2*BigFloat(pi)*im*(1/BigFloat(q)+δ^q))
    x = GetPeriodicPoint(δ, b; z=x0)
    w, dw = dfq(x; λ=λ, b=b)
    return [f(cp(b); λ=λ, b=b)-f(x; λ=λ, b=b), dw-exp(2*BigFloat(pi)*im/q)], x
end
```

```julia
H(δ1, 2, x1)[1]
```

```julia
function dH(δ, b, x0)
    h = BigFloat(1e-200)
    H0, x = H(δ, b, x0)
    H1, x1 = H(δ+h, b, x)
    H2, x2 = H(δ, b+h, x)
    Hx = (H1-H0)/h
    Hy = (H2-H0)/h
    return [Hx Hy]
end
```

```julia
dH(δ1, 2, x1)
```

```julia
function Find(δ, b, x0)
    v = [δ, b]
    x = x0
    for i=1:20
        h, x = H(v[1], v[2], x)
        v -= dH(v[1],v[2],x)\h
    end
    return v, x
end
```
