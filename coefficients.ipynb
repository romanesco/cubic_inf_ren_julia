```julia
using Polynomials, JLD
```

```julia
function trunc_higher_degree(p, degree_limit)
    if (degree(p)<degree_limit)
        return p
    end
    p_coeffs = coeffs(p)  # 係数の配列を取得
    new_coeffs = p_coeffs[1:(degree_limit+1)]  # 指定した次数までの係数を取り出す
    p_truncated = Polynomial(new_coeffs)  # 新しい多項式を作成
    return p_truncated
end
```

```julia
function binaryiterate(g,k,q)
    for i=1:k
        g = trunc_higher_degree(g(g),q+1)
    end
    return truncate(g, atol=1e-5)
end
```
